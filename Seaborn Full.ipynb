## Seaborn

### Distribution plots
- DistPlot
- JoinPlot
- PairPlot

```python
import seaborn as sns
```

```python
df =sns.load_dataset('tips')
df.head()
```

```python
df.shape
```

### Correlation with Heatmap

- A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
df.corr()
```

```python
df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```

```python
sns.heatmap(df.corr(),annot=True,cmap='YlGnBu')
```

```python
sns.heatmap(df.corr(method='spearman'),annot=True,cmap='YlGnBu')
```

### JoinPlot

A join plot allows to study the relationship between 2 numeric variables.<br>The central chart display their correlation.<br>It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='kde')
```

### PairPlot

A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker',kind='scatter',diag_kind='kde')
```

```python
df['smoker'].value_counts()
```

```python
df['time'].value_counts()
```

### DistPlot

Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
sns.distplot(df['total_bill'],kde=False,bins=10)
```

### Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- BoxPlot
- ViolinPlot
- CountPlot
- BarPlot

### CountPlot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='smoker',data=df)
```

### BarPlot

```python
sns.barplot(x='sex',y='tip',data=df)
```

```python
sns.barplot(x='sex',y='tip',data=df,hue='smoker')
```

```python
sns.barplot(x='sex',y='total_bill',data=df,hue='day')
```

### BoxPlot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='tip',data=df)
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.boxplot(x='day',y='total_bill',data=df,hue='sex')
```

### ViolinPlot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x='day',y='total_bill',data=df)
```
