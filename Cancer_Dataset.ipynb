```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
i=load_breast_cancer()
```

```python
dir(i)
```

```python
i.data
```

```python
scaler=MinMaxScaler()
scaler_data=scaler.fit_transform(i.data)
scaler_data
```

```python
i.target
```

```python
i.target_names
```

```python
df=pd.DataFrame(data=scaler_data,columns=i.feature_names)
```

```python
df
```

```python
df.size
```

```python
i.target.size
```

```python
df["target"]=i.target
```

```python
df["target_names"]=[i.target_names[x] for x in i.target]
```

```python
df
```

```python
pd.plotting.scatter_matrix(df[i.feature_names],c=i.target,figsize=(10,10),alpha=.8,marker="o",hist_kwds={"bins":10});
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
DT=DecisionTreeClassifier(max_depth=10,random_state=0)
```

```python
RF=RandomForestClassifier(max_depth=10,random_state=0)
```

```python
svm=SVC(C=1,random_state=0,gamma='scale')
```

```python
loR=LogisticRegression(C=100,random_state=0)
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaler_data,i.target,random_state=0)
```

```python
RF.fit(x_train,y_train)
```

```python
svm.fit(x_train,y_train)
```

```python
loR.fit(x_train,y_train)
```
