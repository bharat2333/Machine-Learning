```python
from sklearn import datasets
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
```

```python
dir(datasets)
```

```python
x,y=make_blobs(random_state=1)
```

```python
len(x),len(y),x.shape,y.shape
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(x)
```

```python
a=kmeans.labels_
a
```

```python
kmeans.predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=a)
```

```python
# Adding more than 3 clusters.
kmeans_5=KMeans(n_clusters=5)
```

```python
kmeans_5.fit(x)
```

```python
b=kmeans_5.labels_
b
```

```python
plt.scatter(x[:,0],x[:,1],c=b)
```

```python
x,y=make_moons(n_samples=200,noise=0.05,random_state=0)
```

```python
len(x),x.shape,len(y),y.shape
```

```python
km_m=KMeans(n_clusters=2)
```

```python
km_m.fit(x)
```

```python
y_pred=km_m.predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(km_m.cluster_centers_[:,0],km_m.cluster_centers_[:,1],c='g',marker='^',s=100, linewidth=2,edgecolor='k')
```

```python
scaler=StandardScaler()
sdata=scaler.fit_transform(x)
```

```python
dbscan=DBSCAN()
```

```python
clusters=dbscan.fit_predict(sdata)
clusters
```

```python
dbscan.labels_
```

```python
plt.scatter(sdata[:,0],sdata[:,1],c=clusters)
```
