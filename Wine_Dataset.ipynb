```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler
```

```python
i=load_wine()
```

```python
dir(i)
```

```python
i.data
```

```python
scaler=MinMaxScaler()
sdata=scaler.fit_transform(i.data)
sdata
```

```python
i.feature_names
```

```python
i.target
```

```python
i.target_names
```

```python
df=pd.DataFrame(data=sdata,columns=i.feature_names)
df
```

```python
df["target"]=i.target
```

```python
df["target_names"]=[i.target_names[x] for x in i.target]
```

```python
df
```

```python
pd.plotting.scatter_matrix(df[i.feature_names],c=i.target,figsize=(15,13),marker="o",hist_kwds={"bins":10});
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
DT=DecisionTreeClassifier(max_depth=3,random_state=0)
```

```python
RF=RandomForestClassifier(max_depth=10,random_state=0)
```

```python
svm=SVC(C=1,random_state=42,gamma='scale')
```

```python
loR=LogisticRegression(C=100,random_state=0)
```

```python
x_train,x_test,y_train,y_test=train_test_split(sdata,i.target,random_state=0)
```

```python
i.data.shape
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
knn.fit(x_train,y_train)
```

```python
DT.fit(x_train,y_train)
```

```python
RF.fit(x_train,y_train)
```

```python
svm.fit(x_train,y_train)
```

```python
loR.fit(x_train,y_train)
```

### Checking accuracy for TEST data

```python
y_test_pred=knn.predict(x_test)
y_test_pred
```

```python
np.mean(y_test_pred==y_test)*100
```

```python
knn.score(x_test,y_test)*100
```

```python
DT.score(x_test,y_test)*100
```
